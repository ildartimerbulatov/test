```python
import pandas as pd
import random

def main():
    lst = ['robot'] * 10 + ['human'] * 10
    random.shuffle(lst)

    data = pd.DataFrame({'whoAmI': lst})

    unique_labels = data['whoAmI'].unique()

    new_data = pd.DataFrame(columns=unique_labels)
    for label in unique_labels:
        new_data[label] = (data['whoAmI'] == label).astype(int)

    print(new_data)

if __name__ == "__main__":
    main()
```
